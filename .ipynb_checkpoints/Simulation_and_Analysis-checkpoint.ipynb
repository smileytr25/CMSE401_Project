```python
from simulation import run_simulation
```

```python
bests = []
best_its = []
rate_rmses = []

for learn_rate in np.arange(0.3, 0.701, 0.01):
    print(f"Learning Rate = {learn_rate:.2f}")

    all_rmses = []

    for _ in range(5):  # run simulation 5 times
        _, _, _, _, _, rmses = run_simulation(learn_rate=learn_rate)
        all_rmses.append(rmses)  # each 'rmses' is a list of RMSEs over 15 iterations

    # Average RMSE for each iteration (i.e., average column-wise)
    all_rmses = np.array(all_rmses)               # shape: (5, 15)
    avg_rmse_per_iter = np.mean(all_rmses, axis=0)  # shape: (15,)

    # Track the best RMSE and its corresponding iteration
    best_rmse = np.min(avg_rmse_per_iter)
    best_iteration = np.argmin(avg_rmse_per_iter)

    print(f"Best RMSE = {best_rmse:.3f} at iteration {best_iteration + 1}\n")

    bests.append(best_rmse)
    best_its.append(best_iteration)
    rate_rmses.append(avg_rmse_per_iter)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 16))
colors = []
ratios = np.array(bests)/np.array(best_its)
for i in range(len(ratios)):
    if i == np.argmin(ratios):
        colors.append("green")
    elif i == np.argmax(ratios):
        colors.append("red")
    else:
        colors.append("yellow")
barplot = ax[0].bar(x=list(np.arange(.3, .71, .01)), height=list(np.array(bests)/np.array(best_its)), ec="black", 
       color=colors, width=.01)
labels = ax[0].set(ylabel="Best RMSE to Best Iteration Ratio", title="RMSE/Iteration Ratio By Learning Rate")
legend = ax[0].legend(handles=[Patch(facecolor="green", label="Minimum Ratio"), Patch(facecolor="red", label="Maximum Ratio")], bbox_to_anchor=(1, .5))

colors = []
min_rate_rmses = rate_rmses[np.argmin(ratios)]
for i in range(len(min_rate_rmses)):
    if i == 0 or min_rate_rmses[i] < min_rate_rmses[i-1]:
        colors.append("green")
    else:
        colors.append("red")
lineplot = ax[1].plot(np.arange(1, 16, 1), rate_rmses[np.argmin(ratios)], color="black")
scatterplot = ax[1].scatter(np.arange(1, 16, 1), rate_rmses[np.argmin(ratios)], ec="black", s=80, zorder=10, c=colors)
ax[1].grid()
legend = ax[1].legend(handles=[Patch(facecolor="green", label="Negative Change"), Patch(facecolor="red", label="Positive Change")], bbox_to_anchor=(1.3, .5))
labels = ax[1].set(ylabel="Mean RMSE Across 5 Trials", title="Mean RMSE vs. Iteration (learning rate = 0.53)")
ax[1].set_ylim((4, 13))
for i, rmse in enumerate(min_rate_rmses):
    x = i + 1
    y = rmse

    # Annotate RMSE value slightly above the point
    ax[1].annotate(f"{y:.3f}",
                   (x, y),
                   textcoords="offset points",
                   xytext=(0, 12),
                   ha='center',
                   fontsize=9,
                   weight='bold',
                   bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=0.5, alpha=0.8))

    # Annotate delta value below the point (if not the first)
    if i > 0:
        delta = y - min_rate_rmses[i - 1]
        ax[1].annotate(f" {delta:+.3f}",
                       (x, y),
                       textcoords="offset points",
                       xytext=(5, 30),
                       ha='center',
                       fontsize=8,
                       color="black",
                       bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="none", alpha=1))
```
